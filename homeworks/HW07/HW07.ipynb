```python
import pandas as pd
from pathlib import Path

# Путь к данным
DATA_DIR = Path("data")

datasets = [
    "S07-hw-dataset-01.csv",
    "S07-hw-dataset-02.csv",
    "S07-hw-dataset-04.csv",
]

dataframes = {}

for fname in datasets:
    print("=" * 100)
    print(f"Датасет: {fname}")
    
    # Загрузка CSV
    df = pd.read_csv(DATA_DIR / fname)
    dataframes[fname] = df
    
    # Первичный анализ
    print("\nhead()")
    display(df.head())
    print("\ninfo()")
    df.info()
    print("\ndescribe() (числовые признаки)")
    display(df.describe())
    
    # Пропуски
    missing_abs = df.isna().sum()
    missing_rel = df.isna().mean()
    
    print("\nПропуски (абсолютное количество)")
    display(missing_abs[missing_abs > 0])
    print("\nПропуски (доля)")
    display(missing_rel[missing_rel > 0])
    
    # Типы признаков
    numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
    categorical_cols = df.select_dtypes(exclude=["number"]).columns.tolist()
    print("\nТипы признаков")
    print(f"Числовые ({len(numeric_cols)}): {numeric_cols}")
    print(f"Категориальные ({len(categorical_cols)}): {categorical_cols}")
    
    # Формирование X и sample_id
    sample_id = df["sample_id"]
    X = df.drop(columns=["sample_id"])
    print("\nПроверка X и sample_id")
    print("X shape:", X.shape)
    print("sample_id shape:", sample_id.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

preprocessors = {}

for fname, df in dataframes.items():
    print("="*100)
    print(f"Препроцессинг для {fname}")
    
    # Разделяем числовые и категориальные признаки
    numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
    categorical_cols = df.select_dtypes(exclude=["number"]).columns.tolist()
    
    # Исключаем sample_id
    if "sample_id" in numeric_cols:
        numeric_cols.remove("sample_id")
    if "sample_id" in categorical_cols:
        categorical_cols.remove("sample_id")
    
    # Препроцессинг числовых признаков
    numeric_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')), 
        ('scaler', StandardScaler())
    ])
    
    # Препроцессинг категориальных признаков 
    if len(categorical_cols) > 0:
        categorical_transformer = Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])
    else:
        categorical_transformer = 'passthrough'  # если категориальных признаков нет

    # ColumnTransformer
    preprocessor = ColumnTransformer([
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
    
    preprocessors[fname] = preprocessor
    print(f"Числовые признаки: {numeric_cols}")
    print(f"Категориальные признаки: {categorical_cols if len(categorical_cols)>0 else 'нет'}")
    print("Препроцессор готов.\n")

# Пример использования для dataset-01
X = dataframes["S07-hw-dataset-01.csv"].drop(columns=["sample_id"])
X_transformed = preprocessors["S07-hw-dataset-01.csv"].fit_transform(X)
print("Форма после трансформации:", X_transformed.shape)


# В датасете 04 присутствуют категориальные признаки.
# Мы осознанно исключаем их из обработки для моделей кластеризации.
# Причина: большинство алгоритмов кластеризации (KMeans, DBSCAN, Agglomerative) работают с числовыми признаками.
# Преобразование категориальных признаков через OneHotEncoder сильно увеличивает размерность,
# что может ухудшить качество кластеризации, особенно на малых выборках или при наличии шумовых признаков.
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# Создаем директорию для графиков
ARTIFACTS_DIR = Path("artifacts/figures/")
ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)

def kmeans_analysis(X, k_range=range(2, 21), random_state=42, dataset_name=""):
    silhouettes = []
    for k in k_range:
        km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
        labels = km.fit_predict(X)
        score = silhouette_score(X, labels)
        silhouettes.append(score)
    
    # График Silhouette vs K
    plt.figure(figsize=(8, 4))
    plt.plot(k_range, silhouettes, marker='o')
    plt.title(f"Silhouette Score vs K (KMeans) - {dataset_name}")
    plt.xlabel("K")
    plt.ylabel("Silhouette Score")
    plt.grid(True)
    
    # Сохраняем график
    fig_name = f"kmeans_silhouette_vs_k_{dataset_name.replace('.csv','').replace('-', '_')}.png"
    plt.savefig(ARTIFACTS_DIR / fig_name, dpi=150, bbox_inches='tight')
    plt.show()
    
    best_k = k_range[np.argmax(silhouettes)]
    print(f"Лучшее K по silhouette: {best_k}")
    km_best = KMeans(n_clusters=best_k, n_init=10, random_state=random_state)
    best_labels = km_best.fit_predict(X)
    return best_labels, km_best

def dbscan_analysis(X, eps_list=[0.5, 1, 1.5], min_samples_list=[3, 5, 10], dataset_name=""):
    best_score = -1
    best_params = None
    best_labels = None
    silhouettes_map = []
    
    for eps in eps_list:
        for min_samples in min_samples_list:
            db = DBSCAN(eps=eps, min_samples=min_samples)
            labels = db.fit_predict(X)
            # Проверяем, что кластеров больше 1
            if len(set(labels)) > 1 and -1 not in set(labels):
                score = silhouette_score(X, labels)
                silhouettes_map.append((eps, min_samples, score))
                if score > best_score:
                    best_score = score
                    best_params = (eps, min_samples)
                    best_labels = labels
    
    # График Silhouette vs eps для min_samples фиксированного
    if silhouettes_map:
        eps_vals, min_samples_vals, scores = zip(*silhouettes_map)
        plt.figure(figsize=(8,4))
        plt.scatter(eps_vals, scores, c=min_samples_vals, cmap='viridis', s=100)
        plt.colorbar(label='min_samples')
        plt.xlabel('eps')
        plt.ylabel('Silhouette Score')
        plt.title(f'Silhouette Score vs eps (DBSCAN) - {dataset_name}')
        plt.grid(True)
        
        # Сохраняем график
        fig_name = f"dbscan_silhouette_vs_eps_{dataset_name.replace('.csv','').replace('-', '_')}.png"
        plt.savefig(ARTIFACTS_DIR / fig_name, dpi=150, bbox_inches='tight')
        plt.show()
    
    if best_params is not None:
        print(f"Лучшие параметры DBSCAN: eps={best_params[0]}, min_samples={best_params[1]}, silhouette={best_score:.3f}")
    else:
        print("DBSCAN не нашел подходящих кластеров (все точки - шум или только один кластер).")
    return best_labels

# --- Основной цикл по датасетам ---
for fname, preprocessor in preprocessors.items():
    print("="*80)
    print(f"Кластеризация, метрики качества и визуализация для {fname}")
    
    X = dataframes[fname].drop(columns=["sample_id"])
    X_transformed = preprocessor.fit_transform(X)
    
    # KMeans
    print("\nKMeans")
    k_labels, k_model = kmeans_analysis(X_transformed, dataset_name=fname)
    print(f"Silhouette Score: {silhouette_score(X_transformed, k_labels):.3f}")
    print(f"Davies-Bouldin Score: {davies_bouldin_score(X_transformed, k_labels):.3f}")
    print(f"Calinski-Harabasz Score: {calinski_harabasz_score(X_transformed, k_labels):.3f}")
    print("Простая статистика по KMeans:", np.bincount(k_labels))
    
    # DBSCAN
    print("\nDBSCAN")
    db_labels = dbscan_analysis(X_transformed, dataset_name=fname)
    if db_labels is not None:
        noise_mask = (db_labels == -1)
        n_noise = noise_mask.sum()
        total = len(db_labels)
        print(f"Доля шума: {n_noise / total:.2%} ({n_noise} точек из {total})")
        
        mask = ~noise_mask
        if mask.sum() > 1:
            X_db = X_transformed[mask]
            labels_db = db_labels[mask]
            print(f"Silhouette Score (non-noise): {silhouette_score(X_db, labels_db):.3f}")
            print(f"Davies-Bouldin Score (non-noise): {davies_bouldin_score(X_db, labels_db):.3f}")
            print(f"Calinski-Harabasz Score (non-noise): {calinski_harabasz_score(X_db, labels_db):.3f}")
            print("Простая статистика по DBSCAN:", np.bincount(labels_db))
        else:
            print("Недостаточно точек без шума для расчёта метрик.")
    else:
        print("DBSCAN не нашел подходящих кластеров.")
    
    # PCA визуализация (2D scatter)
    print("\n--- Визуализация PCA 2D ---")
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_transformed)
    
    # Выбираем "лучший" результат: KMeans (можно вручную менять)
    best_labels = k_labels
    
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=best_labels, cmap='tab10', s=50)
    plt.xlabel('PCA1')
    plt.ylabel('PCA2')
    plt.title(f'PCA 2D scatter для {fname} (по кластерам)')
    plt.legend(*scatter.legend_elements(), title="Кластеры")
    plt.grid(True)
    
    # Сохраняем PCA график
    fig_name = f"pca_scatter_{fname.replace('.csv','').replace('-', '_')}.png"
    plt.savefig(ARTIFACTS_DIR / fig_name, dpi=150, bbox_inches='tight')
    plt.show()

print(f"Графики сохранены в директорию: {ARTIFACTS_DIR}")
```

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans
import numpy as np

# Выбираем первый датасет
fname = "S07-hw-dataset-01.csv"
X = dataframes[fname].drop(columns=["sample_id"])
X_transformed = preprocessors[fname].fit_transform(X)

# Параметры KMeans: используем лучшее K из предыдущего анализа
best_k = KMeans(n_clusters=2, n_init=10, random_state=42).fit(X_transformed).n_clusters  # пример, можно заменить вручную

# 5 запусков с разными random_state
n_runs = 5
labels_list = []

for i in range(n_runs):
    km = KMeans(n_clusters=best_k, n_init=10, random_state=i*10)
    labels = km.fit_predict(X_transformed)
    labels_list.append(labels)

# Считаем матрицу ARI между всеми парами запусков
ari_matrix = np.zeros((n_runs, n_runs))

for i in range(n_runs):
    for j in range(n_runs):
        ari_matrix[i, j] = adjusted_rand_score(labels_list[i], labels_list[j])

print("Матрица ARI между 5 запусков KMeans:")
print(np.round(ari_matrix, 3))

# Средняя устойчивость
upper_tri_indices = np.triu_indices(n_runs, k=1)
mean_ari = ari_matrix[upper_tri_indices].mean()
print(f"\nСреднее ARI между разными запусками: {mean_ari:.3f}")

# Интерпретация:
if mean_ari > 0.9:
    print("Разбиения очень стабильны.")
elif mean_ari > 0.7:
    print("Разбиения относительно стабильны.")
else:
    print("Разбиения нестабильны, KMeans чувствителен к инициализации.")
```

```python
import json
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN

def save_metrics_summary(dataframes, preprocessors, metrics_summary_path):
    """
    Собирает и сохраняет метрики для всех датасетов и моделей в JSON.
    """
    summary = {}
    
    for fname, preprocessor in preprocessors.items():
        print(f"Рассчет метрик для {fname}...")
        X = dataframes[fname].drop(columns=["sample_id"])
        X_transformed = preprocessor.fit_transform(X)

        # --- KMeans ---
        k_range = range(2, 21)
        silhouettes = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes.append(score)
        best_k = k_range[np.argmax(silhouettes)]
        k_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)

        k_silhouette = silhouette_score(X_transformed, k_labels)
        k_davies_bouldin = davies_bouldin_score(X_transformed, k_labels)
        k_calinski_harabasz = calinski_harabasz_score(X_transformed, k_labels)

        # --- DBSCAN ---
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score = -1
        best_params = None
        best_db_labels = None
        
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score:
                        best_score = score
                        best_params = (eps, min_samples)
                        best_db_labels = labels

        if best_db_labels is not None and len(set(best_db_labels)) > 1:
            db_silhouette = silhouette_score(X_transformed, best_db_labels)
            noise_mask = (best_db_labels == -1)
            n_noise = noise_mask.sum()
            total = len(best_db_labels)
            noise_ratio = n_noise / total if total > 0 else 0
            
            mask = ~noise_mask
            if mask.sum() > 1:
                X_db_non_noise = X_transformed[mask]
                labels_db_non_noise = best_db_labels[mask]
                db_davies_bouldin = davies_bouldin_score(X_db_non_noise, labels_db_non_noise)
                db_calinski_harabasz = calinski_harabasz_score(X_db_non_noise, labels_db_non_noise)
            else:
                db_davies_bouldin = None
                db_calinski_harabasz = None
        else:
            db_silhouette = None
            noise_ratio = 1.0
            db_davies_bouldin = None
            db_calinski_harabasz = None
            best_params = None

        summary[fname] = {
            "KMeans": {
                "best_k": int(best_k),
                "silhouette_score": round(k_silhouette, 4),
                "davies_bouldin_score": round(k_davies_bouldin, 4),
                "calinski_harabasz_score": round(k_calinski_harabasz, 4)
            },
            "DBSCAN": {
                "best_params": best_params,
                "silhouette_score": round(db_silhouette, 4) if db_silhouette is not None else None,
                "davies_bouldin_score": round(db_davies_bouldin, 4) if db_davies_bouldin is not None else None,
                "calinski_harabasz_score": round(db_calinski_harabasz, 4) if db_calinski_harabasz is not None else None,
                "noise_ratio": round(noise_ratio, 4)
            }
        }

    os.makedirs(metrics_summary_path.parent, exist_ok=True)
    with open(metrics_summary_path, 'w', encoding='utf-8') as f:
        json.dump(summary, f, indent=4, ensure_ascii=False)
    print(f"Метрики сохранены в {metrics_summary_path}")


def save_best_configs(dataframes, preprocessors, configs_path):
    """
    Сохраняет информацию о лучших параметрах для каждой модели и датасета.
    """
    configs = {}
    
    for fname, preprocessor in preprocessors.items():
        print(f"Поиск лучших конфигураций для {fname}...")
        X = dataframes[fname].drop(columns=["sample_id"])
        X_transformed = preprocessor.fit_transform(X)

        # KMeans: подбор лучшего K
        k_range = range(2, 21)
        silhouettes = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes.append(score)
        best_k = k_range[np.argmax(silhouettes)]

        # DBSCAN: подбор параметров
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score = -1
        best_params = None
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score:
                        best_score = score
                        best_params = {"eps": eps, "min_samples": min_samples}

        k_silhouette_final = max(silhouettes)
        db_silhouette_final = best_score if best_score != -1 else float('-inf')

        if k_silhouette_final >= db_silhouette_final:
            best_model = "KMeans"
            best_config = {"n_clusters": int(best_k)}
            selection_criterion = "highest_silhouette_score"
        else:
            best_model = "DBSCAN"
            best_config = best_params
            selection_criterion = "highest_silhouette_score"

        configs[fname] = {
            "selected_model": best_model,
            "selected_config": best_config,
            "selection_criterion": selection_criterion,
            "kmeans_info": {
                "best_k_by_silhouette": int(best_k),
                "best_silhouette_score": round(k_silhouette_final, 4)
            },
            "dbscan_info": {
                "best_params_by_silhouette": best_params,
                "best_silhouette_score": round(best_score, 4) if best_score != -1 else None
            }
        }

    os.makedirs(configs_path.parent, exist_ok=True)
    with open(configs_path, 'w', encoding='utf-8') as f:
        json.dump(configs, f, indent=4, ensure_ascii=False)
    print(f"Конфигурации сохранены в {configs_path}")


def save_cluster_labels(dataframes, preprocessors, labels_dir):
    """
    Сохраняет CSV-файлы с лейблами кластеров для лучшего решения каждого датасета.
    """
    os.makedirs(labels_dir, exist_ok=True)
    
    for fname, preprocessor in preprocessors.items():
        print(f"Сохранение лейблов для {fname}...")
        X = dataframes[fname].drop(columns=["sample_id"])
        sample_ids = dataframes[fname]["sample_id"]
        X_transformed = preprocessor.fit_transform(X)

        # Подбор K для KMeans
        k_range = range(2, 21)
        silhouettes_k = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes_k.append(score)
        best_k = k_range[np.argmax(silhouettes_k)]
        k_silhouette_final = max(silhouettes_k)

        # Подбор параметров для DBSCAN
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score_db = -1
        best_labels_db = None
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score_db:
                        best_score_db = score
                        best_labels_db = labels

        # Выбор лучшей модели и генерация финальных лейблов
        if k_silhouette_final >= best_score_db:
            final_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)
        else:
            if best_labels_db is not None:
                final_labels = best_labels_db
            else:
                print(f"  -> DBSCAN не нашел кластеры, используем KMeans с K={best_k}.")
                final_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)

        # Сохранение лейблов
        labels_df = pd.DataFrame({
            'sample_id': sample_ids,
            'cluster_label': final_labels
        })
        ds_num = fname.split("-")[-1].split(".csv")[0]
        output_filename = f"labels_hw07_ds{ds_num}.csv"
        output_path = labels_dir / output_filename
        labels_df.to_csv(output_path, index=False)
        print(f"  -> Лейблы сохранены в {output_path}")



# --- Основная ячейка ---
ARTIFACTS_DIR = Path("artifacts/")
METRICS_PATH = ARTIFACTS_DIR / "metrics_summary.json"
CONFIGS_PATH = ARTIFACTS_DIR / "best_configs.json"
LABELS_DIR = ARTIFACTS_DIR / "labels/"
FIGURES_DIR = ARTIFACTS_DIR / "figures/"

save_metrics_summary(dataframes, preprocessors, METRICS_PATH)
save_best_configs(dataframes, preprocessors, CONFIGS_PATH) 
save_cluster_labels(dataframes, preprocessors, LABELS_DIR)

print("\n--- Артефакты эксперимента успешно созданы ---")
print(f"- {METRICS_PATH}")
print(f"- {CONFIGS_PATH}")
print(f"- Директория с лейблами: {LABELS_DIR}")
print(f"- Директория с графиками: {FIGURES_DIR}")
```
